```python
import pandas as pd
```

```python
weather_raw = pd.read_csv('history_forecast_bulk.csv')
```

```python
weather = weather_raw.drop(columns=['forecast dt unixtime','slice dt unixtime','lat','lon','probability','hours'])
```

```python
weather.head()
```

```python
weather.columns
```

```python
weather["forecast dt iso"] = weather["forecast dt iso"].astype(str).str[:20]
weather["slice dt iso"] = weather["slice dt iso"].astype(str).str[:20]
```

```python
weather.head()
```

```python
weather["forecast dt iso"] = pd.to_datetime(weather["forecast dt iso"])
weather["slice dt iso"] = pd.to_datetime(weather["slice dt iso"])
```

```python
weather = weather[weather["forecast dt iso"].dt.hour == 18].sort_values(by=['forecast dt iso','slice dt iso']) #on prend les prévisions de 18h
```

```python
weather = weather[weather["slice dt iso"].dt.date > weather["forecast dt iso"].dt.date] #on prend que les prévisions du lendemain
```

```python
weather.head()
```

```python
weather[["temperature","dew_point","pressure",'ground_pressure','humidity','clouds','wind_speed','wind_deg']] = weather[["temperature","dew_point","pressure",'ground_pressure','humidity','clouds','wind_speed','wind_deg']].rolling(window=2).mean()
weather[['rain','snow','ice','fr_rain','convective']] = weather[['rain','snow','ice','fr_rain','convective']].rolling(window=2).sum()
weather[['snow_depth','accumulated','rate']] = weather[['snow_depth','accumulated','rate']].rolling(window=2).max()
```
